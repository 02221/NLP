```python
score = 0

for i in range(0, news_df.shape[0]):
    
    score = 0
    
    for j in range(0, dic_df.shape[0]):
        
        if dic_df.iloc[j,0] in news_df.iloc[i,1]:
            
            score = score + dic_df.iloc[j,1]
            
    news_df.iloc[i,2] = score
```
